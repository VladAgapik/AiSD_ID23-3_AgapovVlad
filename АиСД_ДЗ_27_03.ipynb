25.2) Класс – десятичная дробь. Методы класса – сумма, разность, произведение, частное десятичных дробей. Поля класса – целая и дробная часть числа.

```python
class DecimalFraction:
    def __init__(self, whole_part, fractional_part):
        self.whole_part = whole_part
        self.fractional_part = fractional_part

    def add(self, other):
        result = DecimalFraction(0, 0)
        result.whole_part = self.whole_part + other.whole_part
        result.fractional_part = self.fractional_part + other.fractional_part
        if result.fractional_part >= 1:
            result.whole_part += 1
            result.fractional_part -= 1
        return result

    def subtract(self, other):
        result = DecimalFraction(0, 0)
        result.whole_part = self.whole_part - other.whole_part
        result.fractional_part = self.fractional_part - other.fractional_part
        if result.fractional_part < 0:
            result.whole_part -= 1
            result.fractional_part += 1
        return result

    def multiply(self, other):
        result = DecimalFraction(0, 0)
        result.whole_part = self.whole_part * other.whole_part
        result.fractional_part = self.fractional_part * other.fractional_part
        return result

    def divide(self, other):
        result = DecimalFraction(0, 0)
        result.whole_part = self.whole_part / other.whole_part
        result.fractional_part = self.fractional_part / other.fractional_part
        return result

# Пример использования класса
fraction1 = DecimalFraction(2, 0.5)
fraction2 = DecimalFraction(1, 0.25)

sum_result = fraction1.add(fraction2)
print(f"Сумма: {sum_result.whole_part} целых и {sum_result.fractional_part} дробных")

diff_result = fraction1.subtract(fraction2)
print(f"Разность: {diff_result.whole_part} целых и {diff_result.fractional_part} дробных")

product_result = fraction1.multiply(fraction2)
print(f"Произведение: {product_result.whole_part} целых и {product_result.fractional_part} дробных")

quotient_result = fraction1.divide(fraction2)
print(f"Частное: {quotient_result.whole_part} целых и {quotient_result.fractional_part} дробных")
```

26.2) Опишите класс Triangle, заданный длинами сторон. Включите в описание класса методы, позволяющие вывести длины сторон треугольника на экран, рассчитать периметр, площадь и высоты треугольника и свойство, позволяющее установить, существует ли треугольник с данными длинами сторон.

```python
import math

class Triangle:
    def __init__(self, side1, side2, side3):
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3

    def display_sides(self):
        print(f"Длины сторон треугольника: {self.side1}, {self.side2}, {self.side3}")

    def calculate_perimeter(self):
        return self.side1 + self.side2 + self.side3

    def calculate_area(self):
        s = self.calculate_perimeter() / 2
        return math.sqrt(s * (s - self.side1) * (s - self.side2) * (s - self.side3))

    def calculate_heights(self):
        area = self.calculate_area()
        height1 = 2 * area / self.side1
        height2 = 2 * area / self.side2
        height3 = 2 * area / self.side3
        return height1, height2, height3

    def is_valid_triangle(self):
        return (self.side1 + self.side2 > self.side3) and (self.side1 + self.side3 > self.side2) and (self.side2 + self.side3 > self.side1)

# Пример использования класса
triangle = Triangle(3, 4, 5)

if triangle.is_valid_triangle:
    triangle.display_sides()
    print(f"Периметр треугольника: {triangle.calculate_perimeter()}")
    print(f"Площадь треугольника: {triangle.calculate_area()}")
    heights = triangle.calculate_heights()
    print(f"Высоты треугольника: {heights[0]}, {heights[1]}, {heights[2]}")
else:
    print("Треугольник с такими сторонами не существует.")
```

Создайте дочерние классы КНИГА (название, фамилия автора, год издания, издательство), СТАТЬЯ (название, фамилия автора, название журнала, его номер и год издания), ЭЛЕКТРОННЫЙ РЕСУРС (название, фамилия автора, ссылка, аннотация) со своими методами вывода информации на экран. Создайте список из  n  изданий, выведите полную информацию из списка, а также организуйте поиск изданий по фамилии автора.

```python
class Publication:
    def __init__(self, title, author_last_name):
        self.title = title
        self.author_last_name = author_last_name

    def display_info(self):
        print(f"Title: {self.title}")
        print(f"Author Last Name: {self.author_last_name}")

    def is_searched_publication(self, search_author_last_name):
        return self.author_last_name == search_author_last_name


class Book(Publication):
    def __init__(self, title, author_last_name, year, publisher):
        super().__init__(title, author_last_name)
        self.year = year
        self.publisher = publisher

    def display_info(self):
        super().display_info()
        print(f"Year: {self.year}")
        print(f"Publisher: {self.publisher}")


class Article(Publication):
    def __init__(self, title, author_last_name, journal_title, journal_number, year):
        super().__init__(title, author_last_name)
        self.journal_title = journal_title
        self.journal_number = journal_number
        self.year = year

    def display_info(self):
        super().display_info()
        print(f"Journal Title: {self.journal_title}")
        print(f"Journal Number: {self.journal_number}")
        print(f"Year: {self.year}")


class ElectronicResource(Publication):
    def __init__(self, title, author_last_name, link, abstract):
        super().__init__(title, author_last_name)
        self.link = link
        self.abstract = abstract

    def display_info(self):
        super().display_info()
        print(f"Link: {self.link}")
        print(f"Abstract: {self.abstract}")


n = 3  # Number of publications
publications = [
    Book("Book 1", "Smith", 2000, "Publisher A"),
    Article("Article 1", "Johnson", "Journal X", 5, 2010),
    ElectronicResource("Resource 1", "Brown", "http://example.com", "Description of the resource")
]

# Display all publications
for publication in publications:
    publication.display_info()
    print()

# Search for publications by author's last name
search_author = "Smith"
print(f"Publications by author with last name '{search_author}':")
for publication in publications:
    if publication.is_searched_publication(search_author):
        publication.display_info()
```
